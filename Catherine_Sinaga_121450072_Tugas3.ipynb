2. Berdasarkan data yang anda pilih tentukan tujuan, user, fungsi dan tone (jika ada) yang sesuai dengan visualisasi yang akan anda buat \\
Tujuan: Identifikasi korelasi antara fitur-fitur seperti Glukosa, Tekanan Darah, BMI, dan hasil diabetes (Outcome). \\
User: Peneliti medis yang memerlukan visualisasi untuk mengeksplorasi hubungan antara faktor kesehatan dan risiko diabetes. \\
Fungsi: Eksplorasi data: Memungkinkan pengguna memahami pola-pola dasar dalam dataset. \\

3. Buatlah visualisasi dengan mengunakan library python seperti matplotlib, seaborn atau tools lain

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the dataset
file_path = '/content/diabetes_null.csv'  # Update this path to your local file path
data = pd.read_csv(file_path)

# Set style for the plots
sns.set(style="whitegrid")

# Plot 1: Distribution of Glucose levels for patients with and without diabetes
plt.figure(figsize=(10,6))
sns.histplot(data=data, x='Glucose', hue='Outcome', multiple='stack', bins=20, palette='Set1')
plt.title('Distribution of Glucose Levels by Diabetes Outcome')
plt.xlabel('Glucose Level')
plt.ylabel('Count')
plt.show()

# Plot 2: Missing data heatmap
plt.figure(figsize=(10,6))
sns.heatmap(data.isnull(), cbar=False, cmap='viridis', yticklabels=False)
plt.title('Heatmap of Missing Data')
plt.show()

# Plot 3: Correlation heatmap of the dataset
plt.figure(figsize=(10,6))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix of Diabetes Dataset')
plt.show()
```
